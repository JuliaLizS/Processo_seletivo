```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Converter o tipo da coluna peso de string para float.

```python
dados['PESO'] = pd.to_numeric(dados['PESO'], errors='coerce')
```

Converter o tipo das colunas DTNASC e DTNASCMAE de int e float para datetime

```python
dados['DTNASC'] = pd.to_datetime(dados['DTNASC'], format='%d%m%Y', errors='coerce')
```

```python
dados['DTNASCMAE'] = pd.to_datetime(dados['DTNASCMAE'], format='%d%m%Y', errors='coerce')
```

```python
dados.info()
```

1- "Crie duas vari√°veis, uma com o c√≥digo e a outra com o nome da UF de nascimento"

```python
uf_ibge = {'11': 'RO', '12': 'AC', '13': 'AM', '14': 'RR', '15': 'PA', '16': 'AP',
           '17': 'TO', '21': 'MA', '22': 'PI', '23': 'CE', '24': 'RN', '25': 'PB',
           '26': 'PE', '27': 'AL', '28': 'SE', '29': 'BA', '31': 'MG', '32': 'ES',
           '33': 'RJ', '35': 'SP', '41': 'PR', '42': 'SC', '43': 'RS', '50': 'MS',
           '51': 'MT', '52': 'GO', '53': 'DF'}
```

```python
dados['COD_UF'] = dados['CODMUNNASC'].astype(str).str[:2]
```

```python
dados['UF'] = dados['COD_UF'].map(uf_ibge)
```

```python
dados.head()
```

"e mostre a distribui√ß√£o."

```python
distribuicao = dados['UF'].value_counts()
```

```python
distribuicao
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x=dados['UF'], order=dados['UF'].value_counts().index)
plt.title('Distribui√ß√£o de UF de Nascimento')
plt.xlabel('UF de Nascimento')
plt.ylabel('Quantidade')
```

3- Quais s√£o as tr√™s menores idades e as tr√™s maiores idades das m√£es?

```python
import datetime
from datetime import datetime
```

```python
data_atual = pd.to_datetime('today')
```

```python
data_atual
```

```python
dados['IDADE_MAE'] = data_atual.year - dados['DTNASCMAE'].dt.year
```

```python
dados.head()
```

```python
tres_menores_idades = dados.sort_values('IDADE_MAE').head(3)
```

```python
tres_menores_idades
```

```python
tres_maiores_idades = dados.sort_values('IDADE_MAE', ascending=False).head(3)
```
