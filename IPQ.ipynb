11- Mostre a distribui√ß√£o de escolaridade de m√£es solteiras, negras e pardas para cada UF

```python
dados.info()
```

* ESCMAE2010 = Escolaridade
* ESTCIVMAE = Estado civil
* Ra√ßa IBGE: (I) Amarelo; (II) Branco; (III) Ind√≠gena; (IV) Pardo e (IV) Preto.
* Estado Civil IBGE: 1 = Solteiro; 2 = Casado; 3 = Vi√∫vo; 4 = Sep. Judicial;
6 = Divorciado; 7 = Desquitado; 9 = Ignorado

```python
dados = dados.dropna(subset=['ESTCIVMAE'])
dados = dados.dropna(subset=['RACACORMAE'])
dados = dados.dropna(subset=['ESCMAE2010'])
```

```python
solteiras_negras_pardas = dados[(dados['ESTCIVMAE'] == 1) & (dados['RACACORMAE'] == 4) | (dados['RACACORMAE'] == 5)]
```

```python
escolaridade_uf = solteiras_negras_pardas.groupby(['UF', 'ESCMAE2010']).size().unstack(fill_value=0)
escolaridade_uf
```

```python
plt.figure(figsize=(10, 6))
escolaridade_uf.plot(kind='barh', stacked=True)
plt.title('Escolaridade das M√£e Solteiras, Negras e Pardas por UF')
plt.xlabel('UF')
plt.ylabel('Quantidade')
plt.legend(title='Escolaridade')
plt.show()
```

12- Qual a porcentagem de mulheres casadas que tiveram parto vaginal com menos de 37 semanas de gesta√ß√£o?

1: Parto Normal
2: Parto Ces√°reo
3: Parto Instrumental
9: N√£o Informado ou Outro


```python
trab_parto = dados.groupby(dados['STCESPARTO']).size()
trab_parto
```

```python
sem_gest = dados.groupby(dados['SEMAGESTAC']).size()
sem_gest
```

```python
dados = dados.dropna(subset=['STCESPARTO'])
dados = dados.dropna(subset=['SEMAGESTAC'])
```

```python
casadas_normal_37 = dados[(dados['ESTCIVMAE'] == 2) & (dados['STCESPARTO'] == 1.0) & (dados['SEMAGESTAC'] < 37.0)]
```

```python
mostrar = casadas_normal_37.groupby(['ESTCIVMAE', 'STCESPARTO', 'SEMAGESTAC']).size().unstack(fill_value=0)
mostrar
```

```python
casadas_normal_37_porcentagem = (len(casadas_normal_37) / len(dados)) * 100
f"{casadas_normal_37_porcentagem:.2f}"
```
